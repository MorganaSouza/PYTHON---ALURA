```python
Aqui está um resumo da aula com exemplos de como usar as funções:
Ordenação básica:
Usar a função sorted(iterable) para retornar uma nova lista ordenada a partir de um iterável:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        idades = [15, 87, 32, 65, 56, 32, 49, 37]
print(sorted(idades))  # [15, 32, 32, 37, 49, 56, 65, 87]
                    
                
        
Usar reversed(iterable) para retornar um iterador que percorre o iterável na ordem inversa:
                 
                    
                        less
                        
                            Copiar código
                        
                    
                    
                        print(list(reversed(idades)))  # [37, 49, 32, 56, 65, 32, 87, 15]
                    
                
        
Combinar sorted() e reversed() para obter uma lista ordenada de forma decrescente:
                 
                    
                        less
                        
                            Copiar código
                        
                    
                    
                        print(list(reversed(sorted(idades))))  # [87, 65, 56, 49, 37, 32, 32, 15]
                    
                
        


Ordenação in-place:
Usar o método sort() da lista para ordenar a lista no próprio local:
                 
                    
                        scss
                        
                            Copiar código
                        
                    
                    
                        idades.sort()
print(idades)  # [15, 32, 32, 37, 49, 56, 65, 87]
                    
                
        
O sort() altera a lista original, enquanto o sorted() retorna uma nova lista ordenada.


Diferenças entre sorted() e sort():
sorted() retorna uma nova lista ordenada, sem alterar a lista original.
sort() ordena a lista no próprio local, alterando a lista original.
A escolha entre sorted() ou sort() depende da necessidade do seu código.


Parâmetros adicionais:
Ambas as funções sorted() e sort() aceitam os parâmetros reverse e key para personalizar a ordenação.
O parâmetro reverse=True inverte a ordem da ordenação.
O parâmetro key permite especificar uma função que será aplicada a cada elemento antes da comparação.



Esses são os principais pontos abordados na aula sobre ordenação de listas no Python. Lembre-se de que a escolha entre sorted() e sort() depende do seu caso de uso específico.
```

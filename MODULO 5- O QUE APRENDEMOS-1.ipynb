```python
# O QUE APRENDEMOS NO MÓDULO 5

BUILDINGS

Claro, vou explicar cada um desses conceitos com exemplos.
Enumerated:
Enumerated é uma função built-in do Python que permite enumerar os elementos de uma sequência, como uma lista ou uma tupla.
Ela retorna um objeto iterável que contém tuplas, onde cada tupla possui dois elementos: o índice (posição) e o valor do elemento correspondente na sequência original.
Exemplo:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        frutas = ['maçã', 'banana', 'laranja', 'uva']
for indice, fruta in enumerate(frutas):
    print(f"Índice: {indice}, Fruta: {fruta}")
                    
                
        Saída:
                 
                    
                        makefile
                        
                            Copiar código
                        
                    
                    
                        Índice: 0, Fruta: maçã
Índice: 1, Fruta: banana
Índice: 2, Fruta: laranja
Índice: 3, Fruta: uva
                    
                
        


Função range:
A função range() é usada para gerar uma sequência de números inteiros.
Ela recebe três argumentos opcionais: início, fim e passo.
O argumento "início" é o primeiro número da sequência (padrão é 0).
O argumento "fim" é o último número da sequência (exclusive).
O argumento "passo" é o incremento entre os números (padrão é 1).
Exemplo:
                 
                    
                        less
                        
                            Copiar código
                        
                    
                    
                        print(list(range(5)))     # [0, 1, 2, 3, 4]
print(list(range(2, 8)))  # [2, 3, 4, 5, 6, 7]
print(list(range(1, 10, 2))) # [1, 3, 5, 7, 9]
                    
                
        


Desempacotar tuplas:
O desempacotamento de tuplas permite atribuir os valores de uma tupla a variáveis separadas.
Isso é útil quando você tem uma sequência de tuplas e deseja acessar os valores individuais de cada tupla.
Exemplo:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        usuarios = [
    ("Guilherme", 37, 1981),
    ("Daniela", 31, 1987),
    ("Paulo", 39, 1979)
]

for nome, idade, nascimento in usuarios:
    print(f"Nome: {nome}, Idade: {idade}, Nascimento: {nascimento}")
                    
                
        Saída:
                 
                    
                        yaml
                        
                            Copiar código
                        
                    
                    
                        Nome: Guilherme, Idade: 37, Nascimento: 1981
Nome: Daniela, Idade: 31, Nascimento: 1987
Nome: Paulo, Idade: 39, Nascimento: 1979
                    
                
        


Função len:
A função len() retorna o comprimento (número de elementos) de um objeto, como uma lista, tupla, string, etc.
Exemplo:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        frutas = ['maçã', 'banana', 'laranja', 'uva']
print(len(frutas))  # Output: 4
                    
                
        



Esses são os principais conceitos que aprendemos nesta aula. O enumerate() nos permite enumerar os elementos de uma sequência, a função range() gera uma sequência de números, o desempacotamento de tuplas facilita o acesso aos valores individuais, e a função len() nos dá o tamanho de um objeto.Espero que esses exemplos tenham ajudado a esclarecer esses conceitos. Se você tiver mais dúvidas, fique à vontade para perguntar.
```

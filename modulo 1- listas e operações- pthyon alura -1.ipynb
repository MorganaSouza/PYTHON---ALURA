Mutabilidade das Listas:
As listas em Python são objetos mutáveis, o que significa que seus elementos podem ser alterados após a sua criação.
Exemplo:
                 
                    
                        makefile
                        
                            Copiar código
                        
                    
                    
                        idades = [16, 21, 29, 56, 43]
idades.append(13)  # Adiciona um novo elemento à lista
idades[0] = 18  # Altera o primeiro elemento da lista
                    
                
        


Passagem de Listas como Parâmetros:
Quando uma lista é passada como parâmetro para uma função, a função recebe uma referência para o mesmo objeto na memória.
Exemplo:
                 
                    
                        scss
                        
                            Copiar código
                        
                    
                    
                        def faz_processamento_de_visualizacao(lista):
    print(len(lista))
    lista.append(13)

idades = [16, 21, 29, 56, 43]
faz_processamento_de_visualizacao(idades)
print(idades)  # Saída: [16, 21, 29, 56, 43, 13]
                    
                
        
Qualquer alteração feita na lista dentro da função irá afetar a lista original.


Valor Padrão Mutável:
Ao definir um valor padrão mutável, como uma lista vazia, para um parâmetro de função, pode ocorrer um comportamento inesperado.
Exemplo:
                 
                    
                        scss
                        
                            Copiar código
                        
                    
                    
                        def faz_processamento_de_visualizacao(lista=[]):
    print(len(lista))
    lista.append(13)

faz_processamento_de_visualizacao()  # Saída: 0
faz_processamento_de_visualizacao()  # Saída: 1
faz_processamento_de_visualizacao()  # Saída: 2
                    
                
        
Cada chamada da função irá utilizar a mesma lista padrão, que será modificada a cada chamada.


Solução para Valor Padrão Mutável:
Para evitar esse problema, a recomendação é usar o valor None como valor padrão e, dentro da função, verificar se o parâmetro é None antes de criar uma nova lista.
Exemplo:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        def faz_processamento_de_visualizacao(lista=None):
    if lista is None:
        lista = []
    print(len(lista))
    lista.append(13)

faz_processamento_de_visualizacao()  # Saída: 0
faz_processamento_de_visualizacao()  # Saída: 0
                    
                
        
Dessa forma, cada chamada da função terá uma nova lista.


Boas Práticas:
Sempre que possível, utilize objetos imutáveis, como tuplas, para evitar problemas de mutabilidade.
Quando for necessário usar objetos mutáveis, como listas, tome cuidado ao passá-los como parâmetros de funções e ao definir valores padrão mutáveis.

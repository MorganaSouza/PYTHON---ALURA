```python
Builtins como enumerated, range e desempacotamento automatico de tuplas

Aqui está um resumo da aula com exemplos de como usar as funções mencionadas:Principais tópicos abordados:
Usando o enumerate() para enumerar os elementos de uma lista:
O enumerate() retorna um objeto iterável que gera tuplas contendo o índice e o valor de cada elemento da lista.
Exemplo:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        idades = [15, 87, 32, 65, 56, 32, 49, 37]
for indice, idade in enumerate(idades):
    print(indice, idade)
                    
                
        Saída:
                 
                    
                        undefined
                        
                            Copiar código
                        
                    
                    
                        0 15
1 87
2 32
3 65
4 56
5 32
6 49
7 37
                    
                
        


Usando o range() para gerar uma sequência de números:
O range() gera uma sequência de números, que pode ser usada em um loop for.
Exemplo:
                 
                    
                        go
                        
                            Copiar código
                        
                    
                    
                        for i in range(len(idades)):
    print(i, idades[i])
                    
                
        Saída:
                 
                    
                        undefined
                        
                            Copiar código
                        
                    
                    
                        0 15
1 87
2 32
3 65
4 56
5 32
6 49
7 37
                    
                
        


Desempacotamento de tuplas:
Quando iteramos sobre uma lista de tuplas, podemos desempacotar os elementos da tupla em variáveis separadas.
Exemplo:
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        usuarios = [
    ("Guilherme", 37, 1981),
    ("Daniela", 31, 1987),
    ("Paulo", 39, 1979)
]
for nome, idade, nascimento in usuarios:
    print(nome)
                    
                
        Saída:
                 
                    
                        undefined
                        
                            Copiar código
                        
                    
                    
                        Guilherme
Daniela
Paulo
                    
                
        
Também é possível ignorar elementos da tupla usando o caractere _.
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        for nome, _, _ in usuarios:
    print(nome)
                    
                
        Saída:
                 
                    
                        undefined
                        
                            Copiar código
                        
                    
                    
                        Guilherme
Daniela
Paulo
                    
                
        



Esses são os principais recursos abordados na aula, que permitem trabalhar de forma mais eficiente com sequências de dados em Python.
```

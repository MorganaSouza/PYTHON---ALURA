```python
# o que aprendemos modulo 6

Claro, vou explicar esses conceitos com exemplos:
Utilizar a função sorted() para fazer a ordenação sem mudar o conteúdo na lista original:


                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        idades = [12, 90, 1, 8, 10, 25, 32, 31, 40, 50]

# Usando o sorted()
idades_ordenadas = sorted(idades)
print(idades_ordenadas)  # Saída: [1, 8, 10, 12, 25, 31, 32, 40, 50, 90]
print(idades)  # A lista original não foi alterada: [12, 90, 1, 8, 10, 25, 32, 31, 40, 50]
                    
                
        Neste exemplo, a função sorted() retorna uma nova lista com os valores ordenados, sem modificar a lista original idades.
Usar a função sort() para ordenar atribuindo e mudando a lista original:


                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        idades = [12, 90, 1, 8, 10, 25, 32, 31, 40, 50]

# Usando o sort()
idades.sort()
print(idades)  # Saída: [1, 8, 10, 12, 25, 31, 32, 40, 50, 90]
                    
                
        Neste caso, o método sort() ordena a lista original idades diretamente, modificando-a.
Utilizar a função reversed() que inverte os valores de uma lista sem alterar a lista original:


                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        idades = [12, 90, 1, 8, 10, 25, 32, 31, 40, 50]

# Usando o reversed()
idades_invertidas = list(reversed(idades))
print(idades_invertidas)  # Saída: [50, 40, 31, 32, 25, 10, 8, 1, 90, 12]
print(idades)  # A lista original não foi alterada: [12, 90, 1, 8, 10, 25, 32, 31, 40, 50]
                    
                
        Neste exemplo, a função reversed() retorna um iterador que percorre a lista original em ordem inversa. Para visualizar os valores invertidos, usamos list(reversed(idades)), que cria uma nova lista com os valores invertidos, sem modificar a lista original idades.Espero que esses exemplos tenham ajudado a entender melhor as diferenças entre as funções sorted(), sort() e reversed() no contexto de listas em Python. Qualquer dúvida, estou à disposição!

```

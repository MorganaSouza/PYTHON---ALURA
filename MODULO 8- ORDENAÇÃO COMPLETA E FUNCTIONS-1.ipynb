```python
#modulo8 ORDENAÇÃO TOTAL 

Aqui está um resumo da aula sobre ordenação completa e functools:Principais pontos:
Quando temos objetos com múltiplos atributos, a ordenação simples pelo saldo não é suficiente. Precisamos de um critério de ordenação mais complexo.

Exemplo:
                 
                    
                        ini
                        
                            Copiar código
                        
                    
                    
                        conta_do_guilherme = ContaSalario(1700, 500)
conta_da_daniela = ContaSalario(3, 1000)
conta_do_paulo = ContaSalario(133, 500)
                    
                
        Neste caso, queremos ordenar primeiro pelo código da conta (id) e depois pelo saldo.
Podemos usar o attrgetter() do módulo functools para criar uma função de ordenação customizada:


                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        from functools import attrgetter

for conta in sorted(contas, key=attrgetter("_codigo", "_saldo")):
    print(conta)
                    
                
        Isso ordena primeiro pelo atributo _codigo e depois pelo _saldo.
Outra opção é implementar os métodos mágicos __lt__ (menor que) e __eq__ (igual a) na classe ContaSalario. Isso permite uma ordenação "natural" dos objetos.


                 
                    
                        ruby
                        
                            Copiar código
                        
                    
                    
                        @total_ordering
class ContaSalario:
    def __lt__(self, outro):
        if self._saldo != outro._saldo:
            return self._saldo < outro._saldo
        return self._codigo < outro._codigo

    def __eq__(self, outro):
        return self._codigo == outro._codigo and self._saldo == outro._saldo
                    
                
        Agora podemos usar o sorted() diretamente nos objetos ContaSalario.
O total_ordering do módulo functools gera automaticamente os métodos __le__, __gt__, __ge__ e __ne__ a partir dos métodos __lt__ e __eq__ implementados.

Exemplo:
                 
                    
                        makefile
                        
                            Copiar código
                        
                    
                    
                        conta_do_guilherme <= conta_da_daniela  # True
conta_do_guilherme < conta_do_paulo     # False
conta_do_guilherme == conta_do_guilherme # True
                    
                
        Portanto, a aula mostrou diferentes formas de implementar uma ordenação completa e customizada para objetos com múltiplos atributos, usando o attrgetter() e a implementação dos métodos mágicos de comparação, com o auxílio do total_ordering do functools.
```

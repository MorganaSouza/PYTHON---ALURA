```python
Implementando o __lt__

Aqui está um resumo da aula sobre ordenação customizada em Python Collections parte 1, com exemplos:Nesta aula, aprendemos como implementar a ordenação personalizada de objetos da classe ContaSalario. O objetivo era permitir que pudéssemos comparar duas contas salariais usando o operador < (menor que).Inicialmente, não era possível comparar diretamente duas ContaSalario, pois elas não implementavam o método especial __lt__ (menor que). Para resolver isso, adicionamos o seguinte método à classe:
                 
                    
                        ruby
                        
                            Copiar código
                        
                    
                    
                        def __lt__(self, outro):
    return self._saldo < outro._saldo
                    
                
        Esse método compara os saldos das duas contas e retorna True se o saldo da conta self for menor que o saldo da conta outro.Com essa implementação, podemos agora comparar diretamente duas ContaSalario usando o operador <:
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        conta_do_guilherme = ContaSalario(500, 'Guilherme')
conta_da_daniela = ContaSalario(1000, 'Daniela')

print(conta_do_guilherme < conta_da_daniela)  # True
print(conta_do_guilherme > conta_da_daniela)  # False
                    
                
        Além disso, podemos usar a função sorted() para ordenar uma lista de ContaSalario de acordo com o critério definido no método __lt__:
                 
                    
                        python
                        
                            Copiar código
                        
                    
                    
                        contas = [
    ContaSalario(500, 'Guilherme'),
    ContaSalario(510, 'Código 133'),
    ContaSalario(1000, 'Código 3')
]

for conta in sorted(contas):
    print(conta)
                    
                
        Esse código irá imprimir as contas ordenadas do menor saldo para o maior:
                 
                    
                        css
                        
                            Copiar código
                        
                    
                    
                        [>>Codigo 17 Saldo 500<<]
[>>Codigo 133 Saldo 510<<]
[>>Codigo 3 Saldo 1000<<]
                    
                
        Portanto, nesta aula, aprendemos a implementar a ordenação personalizada de objetos em Python, utilizando o método especial __lt__ para definir o critério de comparação entre as instâncias da classe ContaSalario.
```

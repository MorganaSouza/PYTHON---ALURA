```python
 Vamos lá:Nesta aula, aprendemos a trabalhar com listas de objetos personalizados, usando a classe ContaCorrente como exemplo. Vamos relembrar os principais pontos:
Criação da classe ContaCorrente:
A classe ContaCorrente possui um construtor __init__ que recebe o código da conta e inicializa o saldo em 0.
Ela também possui um método deposita que aumenta o saldo da conta.
E um método __str__ que retorna uma representação em string do objeto, mostrando o código e o saldo.



Exemplo:
                 
                    
                        ruby
                        
                            Copiar código
                        
                    
                    
                        class ContaCorrente:
    def __init__(self, codigo):
        self.codigo = codigo
        self.saldo = 0
    
    def deposita(self, valor):
        self.saldo += valor
    
    def __str__(self):
        return f"[>>Codigo {self.codigo} Saldo {self.saldo}<<]"
                    
                
        
Criação de objetos ContaCorrente:
Criamos duas contas, uma para o Gui e outra para a Dani.
Depositamos 500 reais na conta do Gui e 1000 reais na conta da Dani.



Exemplo:
                 
                    
                        makefile
                        
                            Copiar código
                        
                    
                    
                        conta_do_gui = ContaCorrente(15)
conta_do_gui.deposita(500)

conta_da_dani = ContaCorrente(47685)
conta_da_dani.deposita(1000)
                    
                
        
Criação de uma lista de contas:
Criamos uma lista chamada contas que contém as referências para as duas contas criadas.
Ao imprimir a lista, vemos que ela mostra os endereços de memória dos objetos.
Usando um loop for, podemos imprimir cada conta individualmente, chamando o método __str__ de cada objeto.



Exemplo:
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        contas = [conta_do_gui, conta_da_dani]
print(contas)  # [<__main__.ContaCorrente object at 0x7fabla6d40>, <__main__.ContaCorrente object at 07fabfla6db70>]

for conta in contas:
    print(conta)
# [>>Codigo 15 Saldo 500<<]
# [>>Codigo 47685 Saldo 1000<<]
                    
                
        
Mutabilidade dos objetos:
Ao criar uma nova lista contas contendo as duas contas e a conta do Gui novamente, percebemos que todas as referências apontam para o mesmo objeto.
Quando depositamos 100 reais na conta do Gui, essa alteração é refletida em todas as referências à essa conta.
Isso acontece porque estamos trabalhando com objetos mutáveis, então precisamos ter cuidado ao manipulá-los dentro de listas.



Exemplo:
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        contas = [conta_do_gui, conta_da_dani, conta_do_gui]
conta_do_gui.deposita(100)

print(contas[0])  # [>>Codigo 15 Saldo 600<<]
print(conta_do_gui)  # [>>Codigo 15 Saldo 600<<]
                    
                
        Esse é o resumo da aula, destacando os principais pontos e exemplos de como usar cada função da classe ContaCorrente e trabalhar com listas de objetos personalizados.
```

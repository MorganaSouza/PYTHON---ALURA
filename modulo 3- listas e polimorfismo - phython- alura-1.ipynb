```python
Aqui está um resumo da aula com exemplos de como usar as funções:Nesta aula, aprendemos sobre herança e polimorfismo em Python, aplicando esses conceitos a uma classe "Conta" e suas subclasses "ContaCorrente" e "ContaPoupanca".Primeiro, criamos a classe "Conta" com atributos privados (_codigo e _saldo) e métodos como deposita() e str():
                 
                    
                        ruby
                        
                            Copiar código
                        
                    
                    
                        class Conta:
    def __init__(self, codigo):
        self._codigo = codigo
        self._saldo = 0

    def deposita(self, valor):
        self._saldo += valor
    
    def __str__(self):
        return "[>>Codigo {} Saldo {}<<]".format(self._codigo, self._saldo)
                    
                
        Em seguida, criamos as subclasses "ContaCorrente" e "ContaPoupanca" que herdam da classe "Conta" e possuem seu próprio método "passa_o_mes()":
                 
                    
                        ruby
                        
                            Copiar código
                        
                    
                    
                        class ContaCorrente(Conta):
    def passa_o_mes(self):
        self._saldo -= 2

class ContaPoupanca(Conta):
    def passa_o_mes(self):
        self._saldo *= 1.01
        self._saldo -= 3
                    
                
        Então, criamos objetos das subclasses, depositamos valores e chamamos o método "passa_o_mes()":
                 
                    
                        bash
                        
                            Copiar código
                        
                    
                    
                        conta16 = ContaCorrente(16)
conta16.deposita(1000)
conta16.passa_o_mes()
print(conta16)  # [>>Codigo 16 Saldo 998<<]

conta17 = ContaPoupanca(17)
conta17.deposita(1000)
conta17.passa_o_mes()
print(conta17)  # [>>Codigo 17 Saldo 1007.0<<]
                    
                
        O polimorfismo é demonstrado quando criamos uma lista de contas e iteramos sobre ela, chamando o método "passa_o_mes()" de cada objeto, independentemente de ser uma "ContaCorrente" ou "ContaPoupanca":
                 
                    
                        makefile
                        
                            Copiar código
                        
                    
                    
                        conta16 = ContaCorrente(16)
conta16.deposita(1000)
conta17 = ContaPoupanca(17)
conta17.deposita(1000)
contas = [conta16, conta17]

for conta in contas:
    conta.passa_o_mes()
                    
                
        Isso funciona porque as subclasses herdam o método "passa_o_mes()" da classe "Conta" e implementam seu próprio comportamento polimórfico.Essa abordagem demonstra os benefícios do polimorfismo, permitindo que trabalhemos com uma sequência de objetos de diferentes tipos, mas com uma interface comum.
```
